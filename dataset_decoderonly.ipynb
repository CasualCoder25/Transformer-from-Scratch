```python
token_list_input = tokenizer.texts_to_sequences(input_sequences_treated)
token_list_output = tokenizer.texts_to_sequences(output_sequences_treated)

def data_loader(index=0):
    # Define block size
    block_size = 50
    count = 0
    # Creating inputs
    encoder_sequence = []
    encoder_padding = []
    decoder_sequence = []
    decoder_padding = []
    decoder_output = []
    padding=np.ones((block_size,))
    while(True):
        for k in range(index,len(token_list_input)):
            for i in range(len(token_list_output[k])):
                for j in range(i+1,i+block_size+1):
                    if j+1<=len(token_list_output[k]):
                        count+=1
                        n_gram_sequence_input_decoder = token_list_output[k][i:j]
                        n_gram_sequence_output_decoder = token_list_output[k][i+1:j+1]
                        n_gram_sequence_padding_decoder = padding
                        n_gram_sequence_input_encoder = token_list_input[k]
                        n_gram_sequence_padding_encoder = padding
                        encoder_sequence.append(n_gram_sequence_input_encoder)
                        encoder_padding.append(n_gram_sequence_padding_encoder)
                        decoder_sequence.append(n_gram_sequence_input_decoder)
                        decoder_padding.append(n_gram_sequence_padding_decoder)
                        decoder_output.append(n_gram_sequence_output_decoder)
                    if count>=32:
                        encoder_sequence_padded=np.array(pad_sequences(
                            encoder_sequence,
                            maxlen=block_size,
                            padding='pre'
                        ))
                        encoder_padding_padded=np.array(pad_sequences(
                            encoder_padding,
                            maxlen=block_size,
                            padding='pre'
                        )).reshape(-1,1,1,block_size)
                        decoder_sequence_padded=np.array(pad_sequences(
                            decoder_sequence,
                            maxlen=block_size,
                            padding='pre'
                        ))
                        decoder_padding_padded=np.array(pad_sequences(
                            decoder_padding,
                            maxlen=block_size,
                            padding='pre'
                        )).reshape(-1,1,1,block_size)
                        decoder_output_padded=np.array(pad_sequences(
                            decoder_output,
                            maxlen=block_size,
                            padding='pre'
                        ))
                        count=0
                        encoder_sequence = []
                        encoder_padding = []
                        decoder_sequence = []
                        decoder_padding = []
                        decoder_output = []
                        yield ((encoder_sequence_padded,encoder_padding_padded,decoder_sequence_padded,decoder_padding_padded),decoder_output_padded)
```
